```python
model = "meta-llama/Llama-2-7b-chat-hf"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model,use_auth_token=True)
```

```python
model = AutoModelForCausalLM.from_pretrained(model,
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True,
                                             load_in_8bit=True,
                                             )
```

```python
pipe  = pipeline("text-generation",
                     model=model,
                     tokenizer=tokenizer,
                     torch_dtype=torch.bfloat16,
                     device_map='auot',
                     max_new_tokens = 512,
                     do_sample=True,
                     top_k=30,
                     num_return_sequences=1,
                     eos_token_id = tokenizer.eos_token_id)
```

```python
llm = HuggingFacePipeline(pipeline=pipe, model_kwargs = {'temperature':0})
```

```python
summary = llm.invoke("Please provide a concise summary of the book Harry Potter")
print(summary)
```

### Initialize the Retrieval QA with Source Chain

```python
from langchain.chains import RetrievalQA
```

```python
query = "Who is Messi"
```

```python
docs = vector_store.similarity_search(query,k=3)
```

```python
qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=docsearch.as_retriever())
```

```python
query = "How good is Vicuna?"
res = qa_chain.run(query)
print(res)
```
